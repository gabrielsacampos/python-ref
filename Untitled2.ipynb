Format Currency:

```python
def format_currency(value, currency="R$"):
    locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
    formatted_value = locale.currency(value, grouping=True)

    if formatted_value.endswith(" R$"):
        formatted_value = f"{currency} " + formatted_value[:-3]  # Remove " R$" do final

    return formatted_value
```
